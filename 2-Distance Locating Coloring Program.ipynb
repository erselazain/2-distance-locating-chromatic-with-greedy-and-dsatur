ALGORITMA GREEDY

```python
from os import wait
import networkx as nx
import numpy as np
import Loc_Col_greedy as lcgd
import time

a = 1
b = 1
m = 10
n = 10
indeks_a = []
indeks_b = []
for i in range(a,m+1):
  indeks_a.append(i)
for i in range(b,n+1):
  indeks_b.append(i)
PK = np.zeros((m,n))
PKL = np.zeros((m,n))
for i,v in enumerate(indeks_a):
  for j,w in enumerate(indeks_b):
     if v < 3 and w < 3 :
        PK[i][j] = 0
        PKL[i][j] = 0
     if v >= 3 and w >= 3 :
        P = nx.path_graph(v)
        K = nx.complete_graph(w)
        pk_kai, pk_kaiL = lcgd.greedy(P,K)
        PK[i][j] = pk_kai
        PKL[i][j] = pk_kaiL
print("PKL = k Pewarnaan Lokasi P tensor K")
print(indeks_b)
print(PKL)
print("PK = k Pewarnaan Jarak 2 P tensor K")
print(indeks_b)
print(PK)
```
